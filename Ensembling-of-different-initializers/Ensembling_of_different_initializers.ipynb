## Importing required libraries

```
import tensorflow as tf
from tensorflow.keras.layers import *
import numpy as np
import matplotlib.pyplot as plt
```

CREATING A HELPING FUNCTION FOR PLOTTING

```
def plotter(history,param):
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.legend(['model_1_loss','model_1_val_loss',
              'model_2_loss','model_2_val_loss',
              'model_3_loss','model_3_val_loss',
              'model_4_loss','model_4_val_loss',
              'model_5_loss','model_5_val_loss'])
  plt.show() 
  plt.savefig(f'Graphs/{param}_loss.png')

  plt.plot(history.history['accuracy'])
  plt.plot(history.history['val_accuracy'])
  plt.xlabel('Epochs')
  plt.ylabel('Accuracy')
  plt.legend(['model_1_acc','model_1_val_acc',
              'model_2_acc','model_2_val_acc',
              'model_3_acc','model_3_val_acc',
              'model_4_acc','model_4_val_acc',
              'model_5_acc','model_5_val_acc'])
  plt.show()
  plt.savefig(f'Graphs/{param}_acc.png')  
```

## Loading the CIFAR-10 Dataset

```
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
x_train , x_test = x_train/255, x_test/255
```

HELPER FUNCTION FOR CREATING THE MODEL

```
def create_model(initializer,activation,name,dropout=0.2):
  model = tf.keras.models.Sequential(name=name)
  model.add(Conv2D(32, (3, 3), activation=activation, kernel_initializer=initializer, padding='same', input_shape=(32, 32, 3)))
  model.add(Conv2D(32, (3, 3), activation=activation, kernel_initializer=initializer, padding='same'))
  model.add(MaxPooling2D((2, 2)))
  model.add(Dropout(0.2))
  model.add(Conv2D(64, (3, 3), activation=activation, kernel_initializer=initializer, padding='same'))
  model.add(Conv2D(64, (3, 3), activation=activation, kernel_initializer=initializer, padding='same'))
  model.add(MaxPooling2D((2, 2)))
  model.add(Dropout(0.3))
  model.add(Conv2D(128, (3, 3), activation=activation, kernel_initializer=initializer, padding='same'))
  model.add(Conv2D(128, (3, 3), activation=activation, kernel_initializer=initializer, padding='same'))
  model.add(MaxPooling2D((2, 2)))
  model.add(Dropout(0.4))
  model.add(Flatten())
  model.add(Dense(128, activation=activation, kernel_initializer=initializer))
  model.add(Dropout(0.5))
  model.add(Dense(10, activation='softmax'))
  model.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
  return model
```

## Training the Models

```
params = {'glorot_uniform':'relu',
          'glorot_uniform':'tanh',
          'he_uniform':'relu',
          'lecun_uniform':'relu'
          }

histories = []

for param in params:
  print("-"*50)
  print(f"Training for {param}")
  for iteration in range(5):
    print(f"Iteration {iteration+1}",end="\n\n")
    model = create_model(param,params[param],name=f'{param+params[param]}_{iteration+1}')
    history = model.fit(x_train,y_train, 512, epochs=30, validation_split=0.2)
    model.save(f'Models/{param}_{iteration+1}.h5')
    histories.append({param:history})
  plotter(history,param)  
```

## Plotting the results

```
for history in histories:
  for h in history:
    # plt.plot(history[h].history['loss'])
    plt.plot(history[h].history['val_loss'])
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
plt.show()
plt.savefig('Graphs/FinalTrainLoss1.png')
for history in histories:
  for h in history:
    # plt.plot(history[h].history['accuracy'])
    plt.plot(history[h].history['val_accuracy'])    
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
plt.show()
plt.savefig('Graphs/FinalTrainAccuracy1.png')    
```

## Analysing the score of category based ensembles

```
from sklearn.metrics import accuracy_score
# testing on non ensembled
for i in range(5):
  glorot_uniform_model = tf.keras.models.load_model(f'Models/glorot_uniform_{i+1}.h5')
  he_uniform_model = tf.keras.models.load_model(f'Models/he_uniform_{i+1}.h5')
  lecun_uniform_model = tf.keras.models.load_model(f'Models/lecun_uniform_{i+1}.h5')

  predictions = np.zeros(y_test.shape)
  p = []
  for i in [glorot_uniform_model,he_uniform_model,lecun_uniform_model]:
    preds = i.predict(x_test)
    predictions = predictions + preds
    preds = np.argmax(preds,1)
    p.append(accuracy_score(y_test,preds))
    

  predictions = predictions/3
  scores = np.argmax(predictions,1)
  print(f"Accuracy for ensembled: {accuracy_score(y_test,scores)}, Accuracies for individual models: {p}")
```

## Analysing ensembling results for combination of multiple initializers

```
predictions = np.zeros(y_test.shape)
for i in range(5):
  model = tf.keras.models.load_model(f'Models/glorot_uniform_{i+1}.h5')
  preds = model.predict(x_test)
  predictions = predictions + preds
predictions = predictions/5
scores = np.argmax(predictions,1)
print(f"Accuracy for Glorot Uniform Ensemble: {accuracy_score(y_test,scores)}")

predictions = np.zeros(y_test.shape)
for i in range(5):
  model = tf.keras.models.load_model(f'Models/he_uniform_{i+1}.h5')
  preds = model.predict(x_test)
  predictions = predictions + preds

predictions = predictions/5
scores = np.argmax(predictions,1)
print(f"Accuracy for He Uniform Ensemble: {accuracy_score(y_test,scores)}")

predictions = np.zeros(y_test.shape)
for i in range(5):  
  model = tf.keras.models.load_model(f'Models/lecun_uniform_{i+1}.h5')
  preds = model.predict(x_test)
  predictions = predictions + preds

predictions = predictions/5
scores = np.argmax(predictions,1)
print(f"Accuracy for Lecun Uniform Ensemble: {accuracy_score(y_test,scores)}")
```
